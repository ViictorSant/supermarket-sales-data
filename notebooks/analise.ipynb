## 1. importação e Carregamento dos Dados

```python

import pandas as pd

df = pd.read_csv("../data/supermarket_sales.csv")
df.head(10)
```

## 2. Tradução das Colunas

As colunas foram renomeadas para facilitar a leitura e interpretação durante a análise.

```python
df = df.rename(columns={
    "Invoice ID": "id_fatura",
    "Branch": "filial",
    "City":"cidade",
    "Customer type": "tipo_cliente",
    "Gender": "genero",
    "Product line": "categoria_produto",
    "Unit price": "preco_unidade",
    "Quantity": "quantidade",
    "Tax 5%": "imposto",
    "Total": "faturamento",
    "Date": "data",
    "Time": "hora",
    "Payment":"pagamento",
    "cogs": "custo_produtos_vendidos",
    "gross margin percentage": "percentual_margem_bruta",
    "gross income": "lucro_bruto",
    "Rating": "avaliacao"
})

df.columns
```

```python
pagamento_map = {
    "Cash": "dinheiro",
    "Ewallet": "carteira_digital",
    "Credit card": "cartao_credito"
}

df["pagamento"] = df["pagamento"].map(pagamento_map)

print(df["pagamento"].unique())
```

## 3. Tradução dos Valores Categóricos

Os valores foram traduzidos para manter a consistência com os nomes das colunas.

```python
map_categoria_produto = {
    "Health and beauty": "saude_beleza",
    "Electronic accessories": "acessorios_eletronicos",
    "Home and lifestyle": "casa_estilo_de_vida",
    "Sports and travel": "esportes_viagem",
    "Food and beverages": "alimentos_bebidas",
    "Fashion accessories": "acessorios_moda"
    }
df["categoria_produto"] = df["categoria_produto"].map(map_categoria_produto)

print(df["categoria_produto"].unique())

```

```python
map_genero = {
    "Female": "feminino",
    "Male": "masculino"
}

df["genero"] = df["genero"].map(map_genero)

print(df["genero"].unique())
```

```python
map_tipo_cliente = {
    "Member": "membro",
    "Normal": "normal"
}

df["tipo_cliente"] = df["tipo_cliente"].map(map_tipo_cliente)

print(df["tipo_cliente"].unique())
```

Valores renomeados:

```python
df.head()
```

## 4. Faturamento por Filial

Ordem decrescente das filiais que mais faturaram.

```python
total_filiais = df.groupby("filial" )["faturamento"].sum()
total_filiais.sort_values(ascending=False, ignore_index=False)
```

A filial **C** lidera o ranking de faturamento total, com pouca distância das filiais **A** e **B**.

## 5. Faturamento de Produtos por Categoria 

Ordem decrescente do faturamento total por categoria de produto

```python
total_categoria_produto = df.groupby("categoria_produto")["faturamento"].sum()
total_categoria_produto.sort_values(ascending=False)
```

A categoria **alimentos e bebidas** lidera o faturamento, **Saúde e beleza** é a menos rentável, com aproximadamente $7.000 de diferença.

## 6. Média do Valor de Compras por Método de Pagamento

Comparação do ticket médio entre as diferentes formas de pagamento.

```python
media_metodo_pagamento = df.groupby("pagamento")["faturamento"].mean()
media_metodo_pagamento.sort_values(ascending=False)
```

A diferença entre os métodos é baixa, indicando que a forma de pagamento não interfere no valor final da compra.

## 7. Faturamento por Filial e Categoria de Produto

Faturamento total, agrupado por filial e categoria, em ordem decrescente.

```python
faturamento_filiais = df.groupby(["filial", "categoria_produto"])["faturamento"].sum()
faturamento_filiais.sort_values(ascending=False)
```

A filial **C** se apresenta como a líder no faturamento por categoria tendo **alimentos e bebidas** como principal destaque.
